```python
import pandas as pd
import numpy as np
```

```python
df_train=pd.read_csv("C:/Users/ayushi/OneDrive/Documents/credit card fraud (train).csv")
```

```python
df_test=pd.read_csv("C:/Users/ayushi/OneDrive/Documents/credit card fraud (test).csv")
```

```python
df_com=pd.concat([df_train,df_test]) #combining both data 
```

```python
df_com['is_fraud'].value_counts()
```

```python
df_com.reset_index(inplace=True)
```

```python
#preprocessing
```

```python
from sklearn.preprocessing import OrdinalEncoder
col = ['trans_date_trans_time', 'merchant', 'category', 'first', 'last',
       'gender', 'street', 'city', 'state', 'job', 'dob', 'trans_num']
one = OrdinalEncoder()
df_com[col] = one.fit_transform(df_com[col])
```

```python
df_com[col]
```

```python
from sklearn.model_selection import train_test_split 
x=df_com.drop(['is_fraud'],axis=1)
y=df_com['is_fraud']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train.shape,x_test.shape,y_test.shape,y_train.shape
```

```python
one
```

```python
#model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
from sklearn.svm import SVC
```

```python
svm=SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred1=log.predict(x_train)
y_pred2=rf.predict(x_train)
y_pred3=svm.predict(x_train)
```

```python
y_pred1
y_pred2
y_pred3
```

```python
print('lr acc : ', accuracy_score(y_train,y_pred1)*100)
print('rf acc : ', accuracy_score(y_train,y_pred2)*100)
print('svm acc : ', accuracy_score(y_train,y_pred3)*100)
```

```python
finaldata=pd.DataFrame({'models':['log','rf','svc'],
             'acc':[accuracy_score(y_train,y_pred1)*100,
                    accuracy_score(y_train,y_pred2)*100,
                    accuracy_score(y_train,y_pred3)*100,]})
```

```python
finaldata #model accuracy
```
